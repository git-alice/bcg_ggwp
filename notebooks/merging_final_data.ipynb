```python
def mapping(series_counties, df, col_name):
    df[col_name] = np.nan
    for county in series_counties.index:
        try:
            city = df.set_index("county").loc[county].city
            #print(city)
            
            df.set_index("city", inplace=True)
            df.loc[city, col_name] = series_counties.loc[county]
            df.reset_index(inplace=True)
        except:
            pass
    return df.set_index("city")
```

```python
df.index.name="city"
df = mapping(life.set_index("Субъект РФ")["оба пола"], df.reset_index(), "life")
```
