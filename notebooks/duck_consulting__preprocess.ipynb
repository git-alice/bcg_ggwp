### Ноутбук обрабатывает и мерджит в один датафрейм данные из ноутбук `duck_consulting__parse`

```python
import pandas as pd
from pathlib import Path

AL = 'А, Б, В, Г, Д, Е, Ё, Ж, З, И, Й, К, Л, М, Н, О, П, Р, С, Т, У, Ф, Х, Ц, Ч, Ш, Щ, Ъ, Ы, Ь, Э, Ю, Я'.split(', ')
```

```python
files = list(Path('./data2/').glob('*'))
len(files)
```

```python
def back_to_norm(s):
    if 'область' in s:
        s = s.replace('область', ' область')
    if 'край' in s:
        s = s.replace('край', ' край')
    for X in AL:
        if X in s[1:]:
            s = s[0] + s[1:].replace(X, f' {X}')
    return s

def get_region_name(path):
    raw_name = path.stem.split('__')[-1]
    return back_to_norm(raw_name)

def get_region_id(path):
    return path.stem.split('__')[0]

def prepare_region_df(path):
    df = pd.read_csv(path)
    df = df.drop(columns=['Unnamed: 0']).dropna(axis=0)
    df = df.iloc[:, [1, 2]]
    df.columns = [0, 1]
    df = df.set_index(df.columns[0]).T
    df.index = [get_region_name(path)]
    return df
```
